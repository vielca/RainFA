```python
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_samples, silhouette_score

from matplotlib import pyplot as plt
import matplotlib.offsetbox
```

```python
results_folder = r'../example/results/silhouette'
stations_parquet_file = r'stations_with_stats.parquet'

if not os.path.exists(results_folder):
    os.mkdir(results_folder)

df = pd.read_parquet(stations_parquet_file)
```

```python
range_n_clusters = [2, 3, 4, 5, 6, 7]

# Possible columns to use with cluster
# 'StationId', 'Name', 'Latitude', 'Longitude', 'Elevation', 'Series Start', 'Series end', 
# 'Min day IR', 'Max day IR', 'Yearly rainfall', 'Number of days with rainfall over 0.2 mm'

df_values = df[['Latitude', 'Longitude', 'Elevation', 'Yearly rainfall', 'Number of days with rainfall over 0.2 mm']]
silhouette_all_list = list()
for n_clusters in range_n_clusters:
    
    clusterer = KMeans(n_clusters=n_clusters)
    cluster_labels = clusterer.fit_predict(df_values)
    df[str(n_clusters)] = cluster_labels

    silhouette_avg = silhouette_score(df_values, cluster_labels)
    silhouette_all_list.append([n_clusters, silhouette_avg])
    print(
        "For n_clusters =",
        n_clusters,
        "The average silhouette_score for all columns is :",
        silhouette_avg,
    )

df.to_excel(os.path.join(results_folder,'silhouette_all_columns.xlsx'), index=False)
df_silhouette_all = pd.DataFrame(silhouette_all_list, columns =['n_clusters', 'silhouette_avg'])


df_values = df[['Latitude', 'Longitude', 'Elevation']]
silhouette_spatial_list = list()
for n_clusters in range_n_clusters:
    
    clusterer = KMeans(n_clusters=n_clusters)
    cluster_labels = clusterer.fit_predict(df_values)
    df[str(n_clusters)] = cluster_labels

    silhouette_avg = silhouette_score(df_values, cluster_labels)
    silhouette_spatial_list.append([n_clusters, silhouette_avg])
    print(
        "For n_clusters =",
        n_clusters,
        "The average silhouette_score for spacial columns is :",
        silhouette_avg,
    )

df.to_excel(os.path.join(results_folder,'silhouette_spacial_columns.xlsx'), index=False)
df_silhouette_spatial = pd.DataFrame(silhouette_spatial_list, columns =['n_clusters', 'silhouette_avg'])


df_values = df[['Yearly rainfall', 'Number of days with rainfall over 0.2 mm']]
silhouette_rain_list = list()
for n_clusters in range_n_clusters:
    
    clusterer = KMeans(n_clusters=n_clusters)
    cluster_labels = clusterer.fit_predict(df_values)
    df[str(n_clusters)] = cluster_labels

    silhouette_avg = silhouette_score(df_values, cluster_labels)
    silhouette_rain_list.append([n_clusters, silhouette_avg])
    print(
        "For n_clusters =",
        n_clusters,
        "The average silhouette_score for rain columns is :",
        silhouette_avg,
    )

df.to_excel(os.path.join(results_folder,'silhouette_rain_columns.xlsx'), index=False)
df_silhouette_rain = pd.DataFrame(silhouette_rain_list, columns =['n_clusters', 'silhouette_avg'])
```

```python
fig, ax = plt.subplots()
rects=ax.bar(df_silhouette_all['n_clusters'],round(df_silhouette_all['silhouette_avg'],3), width=0.8)
ax.bar_label(rects, padding=3)
ax.set_ylabel('Silhouette Width')
ax.set_xlabel('Clusters')
ax.set_title('Silhouette Width')
plt.savefig(os.path.join(results_folder, 'silhouette_all.jpg'))

plt.show()

fig, ax = plt.subplots()
rects=ax.bar(df_silhouette_spatial['n_clusters'],round(df_silhouette_spatial['silhouette_avg'],3), width=0.8)
ax.bar_label(rects, padding=3)
ax.set_ylabel('Silhouette Width')
ax.set_xlabel('Clusters')
ax.set_title('Silhouette Width for spatial data')
plt.savefig(os.path.join(results_folder, 'silhouette_spatial.jpg'))

plt.show()

fig, ax = plt.subplots()
rects=ax.bar(df_silhouette_rain['n_clusters'],round(df_silhouette_rain['silhouette_avg'],3), width=0.8)
ax.bar_label(rects, padding=3)
ax.set_ylabel('Silhouette Width')
ax.set_xlabel('Clusters')
ax.set_title('Silhouette Width for rainfall data')
plt.savefig(os.path.join(results_folder, 'silhouette_rain.jpg'))

plt.show()
```
